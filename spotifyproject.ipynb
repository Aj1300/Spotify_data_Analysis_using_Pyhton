## Exploratory Data Analysis

### Top 5 most popular artist

```python
top_5_artist=df.groupby("artist").count().sort_values(by="song_title",ascending=False)["song_title"][:5]
top_5_artist
```

```python
top_5_artist.plot.bar()
plt.title("Top 5 Most popular artist")
plt.show()
```

### top 5 loudest track

```python
top_5_loundest=df[["loudness","song_title"]].sort_values(by="loudness",ascending=True)[:5]
top_5_loundest
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x="loudness", y="song_title", data=top_5_loundest, marker="o")
plt.title("Top 5 Loudest Songs")
plt.xlabel("Loudness (dB)")
plt.ylabel("Song Title")
plt.grid(True)
plt.show()
```

### artist with most danceble song

```python
top_5_dacanceble_song=df[["danceability","song_title","artist"]].sort_values(by="danceability",ascending=False)[:5]
top_5_dacanceble_song
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability",y="song_title",data=top_5_dacanceble_song)
plt.title("Top 5 dancebility song")
plt.show()
```

## Top 10 Instrumentalness Track


```python
top_10_intrumentalness=df[["instrumentalness","song_title","artist"]].sort_values(by="instrumentalness",ascending=False)[:5]
top_10_intrumentalness
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness",data=top_10_intrumentalness,autopct='%1.2f%%',labels=top_10_intrumentalness.song_title)
plt.title("Top 10 intrumentalness")
plt.show()
```

### Correlation Heatmap between Variable using Pearson correlation method

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr(method='pearson')
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Heatmap between Variables', fontsize=16)
plt.show()
```
