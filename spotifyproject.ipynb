## Exploratory Data Analysis

### Top 5 most popular artist

```python
top_5_artist=df.groupby("artist").count().sort_values(by="song_title",ascending=False)["song_title"][:5]
top_5_artist
```

```python
top_5_artist.plot.barh()
plt.title("Top 5 Most popular artist")
plt.show()
```

### top 5 loudest track

```python
top_5_loundest=df[["loudness","song_title"]].sort_values(by="loudness",ascending=True)[:5]
top_5_loundest
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness",y="song_title",data=top_5_loundest)
plt.title("top 5 loudest")
plt.show()
```

### artist with most danceble song

```python
top_5_dacanceble_song=df[["danceability","song_title","artist"]].sort_values(by="danceability",ascending=False)[:5]
top_5_dacanceble_song
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x="danceability",y="song_title",data=top_5_dacanceble_song)
plt.title("Top 5 dancebility song")
plt.show()
```

## Top 10 Instrumentalness Track


```python
top_10_intrumentalness=df[["instrumentalness","song_title","artist"]].sort_values(by="instrumentalness",ascending=False)[:5]
top_10_intrumentalness
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness",data=top_10_intrumentalness,autopct='%1.2f%%',labels=top_10_intrumentalness.song_title)
plt.title("Top 10 intrumentalness")
plt.show()
```

## Multiple features plot

```python
interest_feature_columns=["acousticness","danceability","duration_ms","energy","instrumentalness","liveness","loudness","speechiness","tempo","valence"]
```

### creating a histogram to know wheather the song was liked

```python
for feature_columns in interest_feature_columns:
  positive_data=df[df["target"]==1][feature_columns]
  negative_data=df[df["target"]==0][feature_columns]

  plt.figure(figsize=(12,7))

  sns.histplot(positive_data,bins=30,label="positive",color="green",kde=True)
  sns.histplot(negative_data,bins=30,label="negative",color="red",kde=True)

  plt.legend(loc="upper right")
  plt.title(f"Postive & negative Histogram Plot for {feature_columns}")
  plt.show()
```
